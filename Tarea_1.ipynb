```python
import pandas as pd
import numpy as np

# Genera Lista_1 con 20 nmeros aleatorios en el intervalo [-1, 1]
Lista_1 = np.random.uniform(-1, 1, 20)

# Genera Lista_2 con 20 nmeros enteros aleatorios en el intervalo [1500, 3000]
Lista_2 = np.random.randint(1500, 3001, 20)

# Crea un diccionario con las listas
data = {"CapSmall": Lista_1, "CapLarge": Lista_2}

# Crea un DataFrame a partir del diccionario
df = pd.DataFrame(data)

# Reindexa los ndices como "Ao 1", "Ao 2", ..., "Ao 20"
df.index = ["Ao {}".format(i) for i in range(1, 21)]

# Realiza estadsticas descriptivas del DataFrame
descripcion = df.describe()

# Obtiene el valor mnimo de la columna "CapSmall"
minimo_CapSmall = df["CapSmall"].min()

# Calcula el promedio de "CapLarge" del ao 10 al ao 20
promedio_CapLarge_10_20 = df["CapLarge"]["Ao 10":"Ao 20"].mean()

# Realiza una grfica de Ao contra "CapLarge"
import matplotlib.pyplot as plt
df["CapLarge"].plot(x=df.index, y="CapLarge", kind="line", title="Ao vs CapLarge")
plt.xlabel("Ao")
plt.ylabel("CapLarge")
plt.show()

# Filtra valores de "CapLarge" mayores a 2000
df_filtrado = df[df["CapLarge"] > 2000]

# Obtiene la dimensin (shape) del nuevo DataFrame filtrado
dimension_df_filtrado = df_filtrado.shape

# Muestra el DataFrame filtrado
print("DataFrame filtrado:")
print(df_filtrado)

# Imprime los resultados
print("\nEstadsticas descriptivas del DataFrame:")
print(descripcion)
print("\nValor mnimo de la columna 'CapSmall':", minimo_CapSmall)
print("\nPromedio de 'CapLarge' del ao 10 al ao 20:", promedio_CapLarge_10_20)
print("\nDimensin del DataFrame filtrado:", dimension_df_filtrado)
```
