```python
import pandas as pd

file_path = "/content/drive/MyDrive/daily_returns.csv"
df = pd.read_csv(file_path)
```

```python
df.set_index('Unnamed: 0', inplace=True)
```

```python
mean = df.mean()
variance = df.var()
```

```python
def transform_column(column):
    mean = column.mean()
    std = column.std()
    return (column - mean) / std
```

```python
df_transformed = df.apply(transform_column)
```

```python
import matplotlib.pyplot as plt

df['AAPL'].plot()
plt.title('Serie de Tiempo de AAPL')
plt.xlabel('Fecha')
plt.ylabel('Valor')
plt.show()
```

```python
import math

def circulo(radio):
  xc= math.pi * (radio ** 2)
  return xc
```

```python
def rectangulo(base, altura):
  xr=base*altura
  return xr
```

```python
import math
def raiz(numero):
  xsqrt=math.sqrt(numero)
  return xsqrt
```
